```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import nltk
import string
import re
```

```python
df = pd.read_csv('Tweets.csv', encoding = 'unicode_escape')
```

```python
print('Dataset size:', df.shape)
print('Dataset are: ', df.columns)
```

```python
def remove_punct(text):
  text = re.sub(r'[^a-zA-Z0-9]', ' ', str(text))
  text = re.sub(r'@[A-Za-z0-9_]+', '', text) 
  text = re.sub(r'\b\w{1,2}\b', '', text)
  text = re.sub(r'\s\s+', ' ', text)
  return text
df['TWEET'] = df['tweet'].apply(lambda x: remove_punct(x))
```

```python
def cleaning_text(text):
  text = re.sub(r'[0]', r' ', text)
  return text
df['author_name'] = df['username'].apply(lambda x: cleaning_text(x))
df.head(5)
```

```python
def tokenization(text):
  text = re.split('\W+', text)
  return text

df['TOKENIZATION'] = df['TWEET'].apply(lambda x: tokenization(x.lower()))
df.head(5)
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(text):
  text = [word for word in text if word not in stopword]
  return text


df['STOP_REMOVAL'] = df['TOKENIZATION'].apply(lambda x: remove_stopwords(x))
df.head(5)
```

```python
df['tweet'] = df['TWEET'].str.lower()
df.head(5)
```

```python
final = df[['tweet_dt', 'username','tweet']]
```
